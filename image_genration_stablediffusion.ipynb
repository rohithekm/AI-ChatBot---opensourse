```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
from IPython.display import display
```

```python
# Load Stable Diffusion model from Hugging Face model hub
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe = pipe.to("cuda") # Ensure the model runs on GPU
#pipe = pipe.to("cpu")  # Ensure the model runs on CPU

# Provide the text prompt for generating an image
prompt = "a fantasy landscape with mountains"
# Generate the image
image = pipe(prompt).images[0]
# Save the generated image
image.save("fantasy_landscape.png")
# displays generated image
display(Image.open("fantasy_landscape.png"))
```

```python
prompt = "a beautiful blue bird sitting on a branch"
image = pipe(prompt).images[0]
image.save("brid_on_branch.png")
display(Image.open("brid_on_branch.png"))
```
